$
\det A = 1 \cdot
\begin{vmatrix}

\end{vmatrix}

\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+ 4 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix}
$

$
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}

$
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}

-1 & 3

$
\det A = 1 \cdot 60 + 2 \cdot 61 + 4 \cdot 27
$

$

$

$

Третья строка:

```python
# рассчитаем корреляцию между столбцами
corr = data.corr()
print('Корреляционная матрица: ')
corr
```

```python
#Сравниваем ранг и столбцы
matrix_rank = np.linalg.matrix_rank(data.select_dtypes(include=[np.number]))
print("Ранг матрицы:", matrix_rank)
num_columns = data.select_dtypes(include=[np.number]).shape[1]
print("Количество столбцов:", num_columns)
```

```python
if matrix_rank < num_columns:

    print("Есть линейная зависимость между некоторыми столбцами")

    # Находим пары столбцов с высокой корреляцией (выше 0.8)
    high_correlation = corr[
        (corr > 0.8) & (corr < 1.0)
    ]
    print("\nСтолбцы с высокой корреляцией:")
    display(high_correlation)

else:
    print("\nНет линейной зависимости между столбцами")
```

```python
X = clean.drop(columns=["cnt"])  # уберем целевую переменную
y = clean["cnt"]
# столбец для смещения (bias)
X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

# находим коэффициенты линейной регрессии
beta = np.linalg.inv(X.T @ X) @ X.T @ y
y_pred = X @ beta
```

```python
# Среднеквадратичная ошибка
mse = np.mean((y - y_pred) ** 2)

# Коэффициент детерминации R^2
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)

print(f"MSE: {mse}")
print(f"R^2: {r2}")
```

Построенная модель линейной регрессии демонстрирует хорошую способность предсказания

R² близко к 0.8 = модель объясняет 79.9%
