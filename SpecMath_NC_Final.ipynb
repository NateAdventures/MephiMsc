##№16

Дана функция:

$
f(x) = \sqrt{\frac{\ln(x^2 - 1)}{x - 2}}
$

Функция определена, если:

1. Подлогарифмическое выражение положительно:

$
x^2 - 1 > 0 \Rightarrow x > 1 \quad \text{или} \quad x < -1
$

2. Знаменатель не равен нулю:

$
x - 2 \neq 0 \Rightarrow x \neq 2
$

3. Корень определён, если выражение внутри него неотрицательно:

$
\frac{\ln(x^2 - 1)}{x - 2} \geq 0
$

Анализируем знак логарифма и дроби. Учитывая, что $\ln(x^2 - 1) > 0$ при $|x| > 1$ и знак дроби меняется в зависимости от $x - 2$, находим:

$
D(f) = (-\infty, -1) \cup (1, 2) \cup (2, \infty)
$

Поскольку логарифм может принимать любые значения, а корень неотрицателен, область значений:

$
E(f) = [0, +\infty)
$

Проверяем $f(-x)$:

$
f(-x) = \sqrt{\frac{\ln((-x)^2 - 1)}{-x - 2}} = \sqrt{\frac{\ln(x^2 - 1)}{- (x + 2)}}
$

Так как знак дроби меняется, получаем $ f(-x) \neq f(x) $ и $ f(-x) \neq -f(x) $.  
Значит, функция \textbf{не является} ни чётной, ни нечётной.

\

1) Область определения: $ (-\infty, -1) \cup (1, 2) \cup (2, \infty) $

2) Область значений: $ [0, +\infty) $

3) Функция не является чётной и не является нечётной.

##№17

Дано:

- $ AC = BC $ (треугольник равнобедренный),
- $ AH $ — высота,
- $ \tan(BAC) = \frac{7}{24} $.

---
Требуется найти $ \cos(BAH) $.

---

\

Вычисляем $ \cos(BAC) $

Используем основное тригонометрическое тождество:

$ \cos(BAC) = \frac{1}{\sqrt{1 + \tan^2(BAC)}} $

Подставляем значение:

$ \cos(BAC) = \frac{1}{\sqrt{1 + \left(\frac{7}{24}\right)^2}} $

\



Вычисляем:

$ \frac{7^2}{24^2} = \frac{49}{576} $

$ 1 + \frac{49}{576} = \frac{576 + 49}{576} = \frac{625}{576} $

$ \sqrt{\frac{625}{576}} = \frac{25}{24} $

$ \cos(BAC) = \frac{24}{25} $

Связь углов $ \angle BAH $ и $ \angle BAC $

Так как $ AH $ — высота в равнобедренном треугольнике, она делит угол $ BAC $ пополам:

$ \angle BAH = \frac{1}{2} \angle BAC $

Используем формулу половинного угла:

$ \cos(BAH) = \sqrt{\frac{1 + \cos(BAC)}{2}} $

---

Вычисляем $ \cos(BAH) $}


$ \cos(BAH) = \sqrt{\frac{1 + \frac{24}{25}}{2}} $

$ = \sqrt{\frac{\frac{25}{25} + \frac{24}{25}}{2}} $

$ = \sqrt{\frac{\frac{49}{25}}{2}} = \sqrt{\frac{49}{50}} $

$ = \frac{7}{\sqrt{50}} = \frac{7\sqrt{2}}{10} $

\

---
Итоговый ответ

\


$ \cos(BAH) = \frac{7\sqrt{2}}{10} $

Ответ: $ \frac{7\sqrt{2}}{10} $.
